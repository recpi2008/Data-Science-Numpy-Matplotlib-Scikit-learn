### Логарифмический масштаб

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 
           'Jupiter', 'Saturn', 'Uranus', 'Neptune']

masses = [0.055274, 0.815, 1.0, 0.107, 
          317.8, 95.0, 14.6, 17.147]
```

```python
plt.plot(planets,masses)
plt.show()
```

```python
#Логарифмический масштаб
plt.plot(planets, masses)
plt.yscale(value="log")

plt.show()
```

## Форматирование текста для графика

```python
rcParams['figure.figsize'] = 6,4.5
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(-5,5,101)
x
```

```python
x2 = x**2
x2
```

```python
plt.plot(x,x2)
plt.show()
```

### Добавим название

```python
plt.plot(x,x2)
plt.title('График f(x)')
plt.show()
```

```python
plt.plot(x,x2)
plt.title('График f(x)',fontsize=16)
plt.show()
```

```python
plt.plot(x,x2)
plt.title('График f(x)',fontsize=16, fontweight='bold')
plt.show()
```

```python
plt.plot(x,x2)
plt.title('График f(x)',fontsize=16, fontweight='bold',color='DarkRed')
plt.show()
```

```python
plt.plot(x,x2)
plt.title('График f(x)',fontsize=16, fontweight='bold',color='#808080')
plt.show()
```

```python
title_font = {
    "fontsize": 16,
    "fontweight": "bold",
    "color": "#808080",
    "family": "serif",
}

plt.plot(x, x2)

plt.title("График функции", fontdict=title_font)

plt.show()
```

```python
plt.plot(x, x2)

plt.title("График функции", fontdict=title_font, loc='left')

plt.show()
```

```python
label_font = {
    "fontsize": 9,
    "family": "serif",
}
```

```python
plt.plot(x, x2)

plt.title("График функции", fontdict=title_font)
plt.xlabel("Переменная x", fontdict=label_font)
plt.ylabel("Функция f(x)", fontdict=label_font)

plt.show()
```

Добавим график еще одной фенкции

```python
x3 = x**3
x3
```

```python
plt.plot(x,x2)
plt.plot(x,x3)

plt.title('Графики f(x^2) f(x^3)', fontdict=title_font)
plt.ylabel('f(x)', fontdict=label_font)
plt.xlabel('Переменная X', fontdict=label_font)

plt.show()
```

### Ограничение осей

```python
plt.axis([-5,5,-8,8])

plt.plot(x,x2)
plt.plot(x,x3)

plt.title('Графики f(x^2) f(x^3)', fontdict=title_font)
plt.ylabel('f(x)', fontdict=label_font)
plt.xlabel('Переменная X', fontdict=label_font)

plt.show()
```

## Объекты библиотеки Matplotlib

```python
x= np.linspace(-5,5,101)
x2= x**2
x3=x**3
```

```python
plt.axis([-5,5,-8,8])
plt.plot(x,x2)
plt.plot(x,x3)
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, x2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8,3.5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8,3.5)
plt.subplots_adjust(wspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3,6)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3,6)
plt.subplots_adjust(hspace=0.3)
```

```python
y = np.sin(x)
z=np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax2.plot(x, x3)
ax3.plot(x, y)
ax4.plot(x, z)

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('X^2')
ax2.plot(x, x3)
ax2.set_title('X^3')
ax3.plot(x, y)
ax3.set_title('X^3')
ax4.plot(x, z)
ax4.set_title('X^4')

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, sharey=True)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('X^2')
ax1.set_xlim([-3,3])

ax2.plot(x, x3)
ax2.set_title('X^3')

ax3.plot(x, y)
ax3.set_title('X^3')

ax4.plot(x, z)
ax4.set_title('X^4')

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, sharey=True)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, x2)
ax1.set_title('X^2')
ax1.set_xlim([-3,3])

ax2.plot(x, x3)
ax2.set_title('X^3')

ax3.plot(x, y)
ax3.set_title('X^3')
ax1.set_ylim([-25,25])

ax4.plot(x, z)
ax4.set_title('X^4')

fig.set_size_inches(8,6)
plt.subplots_adjust(wspace=0.3,hspace=0.3)
```
